```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['car name']=le.fit_transform(data['car name'])
```

```python
data
```

```python
(data == '?').sum()
```

```python

data.replace('?', np.nan, inplace=True)
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
```

```python
correlation_data=data.corr()
```

```python
correlation_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('car name',axis=1)
y=data['car name']
```

```python
y
```

```python
y.head()
```

```python
X
```

```python
X = X.dropna()
y = y[X.index]
X = X.dropna(axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
R_model=Ridge(alpha=1.0)
```

```python
X = X.fillna(X.mean())
```

```python
R_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = R_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
import seaborn as sns
X_test_one_feature = X_test['horsepower']  # replace with your actual column name
plt.figure(figsize=(7, 5))
sns.regplot(x=X_test_one_feature, y=y_pred, color='blue')
plt.xlabel('Horsepower')
plt.ylabel('Predicted MPG')
plt.title('Regression Plot: Horsepower vs Predicted MPG')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(R_model, X_train, y_train,cv=5)
print('cross_val_score;',scores)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

model = RandomForestRegressor(random_state=42)
scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print("Random Forest CV R²:", scores.mean())
```

```python
model = RandomForestRegressor(random_state=42)
model.fit(X, y)
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.show()
```
