```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X = data.drop('Close', axis=1)  # All columns except 'Close'
y = data['Close']               # Only 'Close' as the target
```

```python
#spliting in Training and testing
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
#declear randomforestfregressor as RFR
RFR=RandomForestRegressor()
```

```python
#now Fit X_trai and y_train in RFR
RFR.fit(X_train,y_train)
```

```python
#predict X_test
y_pred=RFR.predict(X_test)
```

```python
#import all these function from metrics
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
#show Mean squared error
print("mean_absolute_error:",mean_squared_error(y_pred,y_test))
```

```python
#import GridSearchCv for improvement
from sklearn.model_selection import GridSearchCV
```

```python
#declear functions of RFR and it in param_grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt']
}
```

```python
#now store GridSearchCv in gridSearchCv
grid_search = GridSearchCV(
    estimator=RFR,
    param_grid=param_grid,
    cv=5,                    # 5-fold cross-validation
    scoring='neg_mean_absolute_error',
    n_jobs=-1,               # Use all cores
    verbose=2
)
```

```python
#fit int GridSearchCV
grid_search.fit(X_train,y_train)
```

```python
#put all thest things in Best model variable
best_model = grid_search.best_estimator_
#predict X_test through GridSearchCV
y_pred = best_model.predict(X_test)
#find Avarge data
mae = mean_absolute_error(y_test, y_pred)
#This line prints the best hyperparameters that were found by GridSearchCV during the model tuning process
print("Best Parameters:", grid_search.best_params_)
#just print MAE
print("Improved MAE:", mae)
#It calculates and prints the R² score (coefficient of determination) for your model’s predictions
print('r2_score:',r2_score(y_test,y_pred))
#These two lines evaluate your model’s prediction errors using Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)
print('mean squared error:',mean_squared_error(y_test,y_pred))
print('root mean squared error;',root_mean_squared_error(y_test,y_pred))
```
