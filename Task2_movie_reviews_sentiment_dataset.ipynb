```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report
```

```python
data=pd.read_csv(f'movie_reviews_sentiment_dataset.csv')
pd.set_option('display.max_rows',None)
data
```

```python
data.head()
```

```python
data.columns
```

```python
data['Review']=data['Review'].str.lower()
```

```python
import re
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords 
```

```python
stopwords=set(stopwords.words('english'))
```

```python
def clean_text(text):
    text = re.sub(r"[^a-zA-Z]", " ", text)  # Remove non-letters
    words = text.lower().split()            # Lowercase and split into words
    words = [w for w in words if w not in stopwords]  # Remove stopwords
    return " ".join(words)
data["Review"] = data["Review"].apply(clean_text)
```

```python
X_train,X_test,y_train,y_test=train_test_split(data["Review"],data["Sentiment"],test_size=0.25,random_state=42)
```

```python
vector=CountVectorizer()
```

```python
X_train_vec=vector.fit_transform(X_train)
X_test_vec=vector.transform(X_test)
```

```python
#LogisticRegression 
lr_model=LogisticRegression()
lr_model.fit(X_train_vec,y_train)
```

```python
nb_model=MultinomialNB()
nb_model.fit(X_train_vec,y_train)
```

```python
y_pred_lr=lr_model.predict(X_test_vec)
y_pred_nb=nb_model.predict(X_test_vec)
```

```python
print("LogisticRegression:",classification_report(X_test,y_pred_lr))
print("Naive Bayes:",classification_report(X_test,y_pred_nb))
```

```python
import joblib
joblib.dump(vector, 'vector.pkl')
joblib.dump(nb_model, 'naive_bayes_model.pkl')
joblib.dump(lr_model, 'logistic_regression_model.pkl')
```
