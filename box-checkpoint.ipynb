```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd. DataFrame({"value" : [54,32,71,38,49,84,29,38,119]})
```

```python
data==data.plot.box()
```

```python
data.plot.box()
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Boxplot with pandas")
plt.show()
```
